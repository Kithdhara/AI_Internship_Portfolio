```python
# 1. Import the visual tools
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import os

# 2. Set the "style" to make charts look professional
sns.set_theme(style="whitegrid")

# 3. Load the CLEAN data (Moving up one folder to find 'data')
data_path = os.path.join("..", "data", "insurance_cleaned.csv")

if os.path.exists(data_path):
    df = pd.read_csv(data_path)
    print("✅ Data Loaded Successfully!")
    print(f"   Shape: {df.shape}")
    # Show the first 5 rows to check
    display(df.head())
else:
    print("❌ Error: File not found. Make sure 'data/insurance_cleaned.csv' exists.")
```

```python
# Create a figure
plt.figure(figsize=(10, 6))

# Draw the Box Plot
# x = Groups (Smoker Yes/No)
# y = The value we are measuring (Charges)
sns.boxplot(data=df, x='smoker', y='charges', palette='Set2')

# Add titles (Always label your charts!)
plt.title("Medical Insurance Costs: Smokers vs. Non-Smokers", fontsize=16)
plt.xlabel("Is Smoker?", fontsize=12)
plt.ylabel("Charges ($)", fontsize=12)

# Show the plot
plt.show()
```

```python
# Create a figure
plt.figure(figsize=(10, 6))

# Scatter Plot
# x = BMI (Body Mass Index)
# y = Cost
# hue = Color dots by 'smoker' (Blue for No, Orange for Yes)
sns.scatterplot(data=df, x='bmi', y='charges', hue='smoker', palette='magma', alpha=0.7)

# Titles
plt.title("Impact of BMI on Insurance Costs (Smokers vs. Non-Smokers)", fontsize=16)
plt.xlabel("BMI", fontsize=12)
plt.ylabel("Charges ($)", fontsize=12)

plt.show()
```
